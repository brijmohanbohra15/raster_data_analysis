```python
# Read band 2 of Landsat-8 for Dehradun region
band1 = gdal.Open('LC08_B1.TIF_mumbai.TIF')
band2 = gdal.Open('LC08_B2.TIF_mumbai.TIF')
band3 = gdal.Open('LC08_B3.TIF_mumbai.TIF')
band4 = gdal.Open('LC08_B4.TIF_mumbai.TIF')
band5 = gdal.Open('LC08_B5.TIF_mumbai.TIF')
band6 = gdal.Open('LC08_B6.TIF_mumbai.TIF')
band7 = gdal.Open('LC08_B7.TIF_mumbai.TIF')
band9 = gdal.Open('LC08_B9.TIF_mumbai.TIF')
band10 = gdal.Open('LC08_10.TIF_mumbai.TIF')
band11 = gdal.Open('LC08_11.TIF_mumbai.TIF')
```

```python
# Read the band as array for manipulation
band1Arr = band1.ReadAsArray().astype('float')
band2Arr = band2.ReadAsArray().astype('float')
band3Arr = band3.ReadAsArray().astype('float')
band4Arr = band4.ReadAsArray().astype('float')
band5Arr = band5.ReadAsArray().astype('float')
band6Arr = band6.ReadAsArray().astype('float')
band7Arr = band7.ReadAsArray().astype('float')
band9Arr = band9.ReadAsArray().astype('float')
band10Arr = band10.ReadAsArray().astype('float')
band11Arr = band11.ReadAsArray().astype('float')
```

```python
band1Arr[band1Arr==0] = np.nan
xmin = np.nanmin(band1Arr)
xmax = np.nanmax(band1Arr)
fig, ax = plt.subplots(figsize =(10,8))
plt.hist(x= band1Arr.flatten(), color ='purple',bins=100)
ax.set(title="Distribution of DN values of band 1",
       xlabel = "DN values",
       ylabel ="Frequency",
      xlim = (xmin,xmax)
      )
plt.show()
```

```python
cal = gdal.Open('L5_Mumbai_cal.tif')
# Read the band as array for manipulation
calArr = cal.ReadAsArray().astype('float')
```

```python
plt.imshow(calArr)
plt.colorbar()
```

```python
# Detect rows of line drop out
index = np.where(calArr==0)
row = index[0][0]
row
```

####  Replace line drop out by the average of the top and bottom line

```python
top =index[0][0]-1
bottom =  index[0][0] +1
top, bottom
```

```python
calArr[row]= (calArr[top] + calArr[bottom])/2
calArr[row]
```
